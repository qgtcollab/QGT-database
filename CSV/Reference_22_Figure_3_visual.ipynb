```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_excel('https://github.com/prokudin/PSU_PHYS496/raw/master/database/expdata/Ref_22_figure_3_obsv_1.xlsx')
df_2 = pd.read_excel('https://github.com/prokudin/PSU_PHYS496/raw/master/database/expdata/Ref_22_figure_3_obsv_2.xlsx')
df_3 = pd.read_excel('https://github.com/prokudin/PSU_PHYS496/raw/master/database/expdata/Ref_22_figure_3_obsv_3.xlsx')
```

```python
energy_bin = 5.55
t_bin = (-0.4, -0.33) # (t_n, t_d), n = neutron, d = deuteron

bin_to_plot = "t_n == " + str(t_bin[0])+ " and t_d == " + str(t_bin[1])+ " and energy == " + str(energy_bin)
```

```python
plot_df_1 = df_1.query(bin_to_plot)
plot_df_2 = df_2.query(bin_to_plot)
plot_df_3 = df_3.query(bin_to_plot)
```

```python
plot_df_1
```

```python
plot_df_2
```

```python
plot_df_3
```

```python
# Generates 3 plots, each associated with a single measurement from the bin specified above

fig_1, ax_1 = plt.subplots(figsize = (10, 8), constrained_layout = True)
fig_2, ax_2 = plt.subplots(figsize = (10, 8), constrained_layout = True)
fig_3, ax_3 = plt.subplots(figsize = (10, 8), constrained_layout = True)
ax_1.errorbar(plot_df_1.phi, plot_df_1.value,yerr = (np.sqrt(plot_df_1.stat_u**2 + plot_df_1.syst_u_low**2), np.sqrt(plot_df_1.stat_u**2 + plot_df_1.syst_u_high**2)), fmt='o', c='k')
ax_2.errorbar(plot_df_2.phi, plot_df_2.value,yerr = (np.sqrt(plot_df_2.stat_u**2 + plot_df_2.syst_u_low**2), np.sqrt(plot_df_2.stat_u**2 + plot_df_2.syst_u_high**2)), fmt='o', c='b')
ax_3.errorbar(plot_df_3.phi, plot_df_3.value,yerr = (np.sqrt(plot_df_3.stat_u**2 + plot_df_3.syst_u_low**2), np.sqrt(plot_df_3.stat_u**2 + plot_df_3.syst_u_high**2)), fmt='o', c='r')
ax_1.tick_params(direction='in',labelsize=15)
ax_2.tick_params(direction='in',labelsize=15)
ax_3.tick_params(direction='in',labelsize=15)
ax_1.text(0.45,0.90,r'$\rm JLab$',transform=ax_1.transAxes,size=20)
ax_1.text(0.45,0.80,r'$t_n~(GeV^2) = %.2f$'%(t_bin[0]),transform=ax_1.transAxes,size=20)
ax_1.text(0.45,0.70,r'$t_d~(GeV^2) = %.2f$'%(t_bin[1]),transform=ax_1.transAxes,size=20)
ax_1.text(0.45,0.60,r'$obsv = d^4\sigma_n + rd^4\sigma_d$',transform=ax_1.transAxes,size=20)
ax_2.text(0.45,0.90,r'$\rm JLab$',transform=ax_2.transAxes,size=20)
ax_2.text(0.45,0.80,r'$t_n~(GeV^2) = %.2f$'%(t_bin[0]),transform=ax_2.transAxes,size=20)
ax_2.text(0.45,0.70,r'$t_d~(GeV^2) = %.2f$'%(t_bin[1]),transform=ax_2.transAxes,size=20)
ax_2.text(0.45,0.60,r'$obsv = d^4\sigma_n$',transform=ax_2.transAxes,size=20)
ax_3.text(0.45,0.90,r'$\rm JLab$',transform=ax_3.transAxes,size=20)
ax_3.text(0.45,0.80,r'$t_n~(GeV^2) = %.2f$'%(t_bin[0]),transform=ax_3.transAxes,size=20)
ax_3.text(0.45,0.70,r'$t_d~(GeV^2) = %.2f$'%(t_bin[1]),transform=ax_3.transAxes,size=20)
ax_3.text(0.45,0.60,r'$obsv = d^4\sigma_d$',transform=ax_2.transAxes,size=20)
ax_1.set_xlabel(r'$\phi~({\rm degree})$', size=20)
ax_1.set_ylabel(r'${d^4\sigma}~({\rm nb/GeV^4})$', size=20)
ax_2.set_xlabel(r'$\phi~({\rm degree})$', size=20)
ax_2.set_ylabel(r'${d^4\sigma}~({\rm nb/GeV^4})$', size=20)
ax_3.set_xlabel(r'$\phi~({\rm degree})$', size=20)
ax_3.set_ylabel(r'${d^4\sigma}~({\rm nb/GeV^4})$', size=20)
```
