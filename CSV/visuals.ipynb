```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#df = pd.read_excel('10001.xlsx')
df = pd.read_excel('Ref_18_Table_1.xlsx')
```

```python
df
```

```python
# choose the bin to plot
x_bin  = 0.126
Q2_bin = 1.11
t_bin  = 0.11
bin_to_plot = "x == " + str(x_bin)+ " and Q2 == " + str(Q2_bin)+ " and t == " + str(t_bin)
```

```python
plot_df = df.query(bin_to_plot)
```

```python
plot_df
```

```python
fig, ax = plt.subplots(figsize = (10, 8), constrained_layout = True)
ax.errorbar(plot_df.phi, plot_df.value,yerr=np.sqrt(plot_df.stat_u**2+ plot_df.syst_u**2),fmt='o',c='k')
ax.tick_params(direction='in',labelsize=15)
ax.text(0.75,0.25,r'$\rm '+plot_df.col[0]+'$',transform=ax.transAxes,size=20)
ax.text(0.75,0.2,r'$x_{Bj} = %.3f$'%(x_bin),transform=ax.transAxes,size=20)
ax.text(0.75,0.15,r'$-t (GeV^2) = %.2f$'%(t_bin),transform=ax.transAxes,size=20)
ax.text(0.75,0.1,r'$Q^2 (GeV^2) = %.2f$'%(Q2_bin),transform=ax.transAxes,size=20)
ax.set_xlabel(r'$\phi~({\rm degree})$',size=20)
ax.set_ylabel(r'${d^4\sigma}~({\rm nb/GeV^4})$',size=30) 
```
